```python
import pandas as pd
import numpy as np
```

```python
alternative_data = pd.read_csv('data/cleaned_alternative_data.csv', index_col='Date')
technical_data = pd.read_csv('data/ta_df.csv', index_col='Date')
```

```python
full_df = alternative_data.join(technical_data)
```

```python
full_df.head()
```

```python
full_df.tail()
```

### Data Cleaning

```python
full_df.replace([np.inf, -np.inf], np.nan, inplace=True)
full_df.dropna(inplace=True)
```

```python
#Remove outliers
full_df = full_df[(full_df['EMA12_pctchg'] < 3.0) & (full_df['EMA12_pctchg'] > -3.0)]
full_df = full_df[(full_df['Pct_Chg'] < 10.0) & (full_df['Pct_Chg'] > -10.0)]
```
